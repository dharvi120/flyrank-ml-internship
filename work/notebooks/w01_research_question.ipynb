# Content Refresh Prioritization

I chose the **Content Refresh Prioritization** lane because content teams often have thousands of pages but limited time to update them. ## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

# Research Question

## Decision

Which content pages should be prioritized for refresh before their search performance declines further?

## Action

The SEO or content management team reviews the highest-ranked pages, updates their content, and republishes them.

## Cost of a Wrong Recommendation

If truly declining pages are not prioritized, they may continue losing search visibility and traffic.

If healthy pages are incorrectly recommended for refresh, the team spends time and effort on pages that do not need immediate attention.

## Why Machine Learning?

Many observable signals influence page performance simultaneously, including content age, search impressions, average ranking position, click-through rate, and update history.

Because these relationships are difficult to capture using simple manual rules, machine learning can help rank pages more effectively and provide better decision support.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Dataset Shape:", df.shape)

corr = df["search_volume"].corr(df["impressions_90d"])
print("Correlation between Search Volume and Impressions:", round(corr, 3))

median_wc = df.groupby("trend_direction")["word_count"].median()
print("\nMedian Word Count by Trend Direction")
print(median_wc)

print("\nUnique Trend Categories:")
print(df["trend_direction"].value_counts())
```

# Observations

The starter dataset contains **30,000 content pages**, providing enough examples for exploratory analysis and model development.

The correlation between **search volume** and **page impressions** is approximately **0.001**, indicating that keyword search volume alone is not a reliable predictor of actual page traffic.

Median word counts are similar across growing and declining pages, suggesting that simply increasing content length is unlikely to solve declining performance.

These observations indicate that multiple features should be considered together when prioritizing pages for refresh.
